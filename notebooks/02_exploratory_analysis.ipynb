```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../outputs/cleaned_crime.csv")

# Quick overview
display(df.describe())
print("Missing values:\n", df.isna().sum())

# Per-capita crime rate per 100k
df["crime_rate_per_100k"] = (df["total_crimes"] / (df["population_lakhs"] * 100000)) * 100000

# Top 10 states in 2022
top_states = df[df["year"] == 2022].sort_values("crime_rate_per_100k", ascending=False).head(10)

plt.figure(figsize=(10,6))
sns.barplot(data=top_states, x="crime_rate_per_100k", y="state_ut", palette="viridis")
plt.title("Top 10 States by Crime Rate (2022)")
plt.xlabel("Crime rate per 100k population")
plt.ylabel("State/UT")
plt.show()
```
