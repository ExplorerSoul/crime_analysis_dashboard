```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

df = pd.read_csv("../outputs/cleaned_crime.csv")

# Trend of total crimes over years
plt.figure(figsize=(12,6))
sns.lineplot(data=df.groupby("year")["total_crimes"].sum().reset_index(),
             x="year", y="total_crimes", marker="o")
plt.title("Total Crimes Trend in India (2020-2022)")
plt.show()

# Heatmap of crimes by state & year
pivot = df.pivot(index="state_ut", columns="year", values="total_crimes")
plt.figure(figsize=(14,10))
sns.heatmap(pivot, cmap="Reds", annot=False)
plt.title("Heatmap of Total Crimes by State & Year")
plt.show()
```
